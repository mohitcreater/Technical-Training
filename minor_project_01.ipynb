##### Public Library Visitor Timing Analysis ######


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {
    "visitor_id": list(range(1, 51)),
    "entry_time": [
        "08:30","08:45","09:00","09:05","09:15","09:25","09:40","09:55",
        "10:00","10:05","10:15","10:20","10:30","10:45","10:55",
        "11:00","11:10","11:20","11:35","11:50",
        "12:00","12:10","12:25","12:40","12:55",
        "13:00","13:10","13:25","13:40","13:55",
        "14:05","14:15","14:30","14:45",
        "15:00","15:10","15:25","15:40","15:55",
        "16:00","16:10","16:25","16:40","16:55",
        "17:05","17:15","17:30","17:45",
        "18:00","18:15"
    ]
}

df = pd.DataFrame(data_dict)
```

```python
df.to_csv("visitor_data.csv", index=False)
print("Dataset created successfully!\n")
```

```python
data = pd.read_csv("visitor_data.csv")
print("First 5 records:\n")
print(data.head())
```

```python
data['entry_time'] = pd.to_datetime(data['entry_time'], format='%H:%M')
data['entry_hour'] = data['entry_time'].dt.hour
```

```python
print("\nDescriptive Statistics:")
print(data['entry_hour'].describe())
```

```python
peak_hour = data['entry_hour'].mode()[0]
print(f"\nPeak Usage Hour: {peak_hour}:00 - {peak_hour}:59")
```

```python
hourly_visits = data['entry_hour'].value_counts().sort_index()
print("\nVisitors Per Hour:")
print(hourly_visits)
```

```python
plt.figure()
plt.hist(data['entry_hour'], bins=range(0, 25), edgecolor='black')
plt.xlabel("Hour of Entry")
plt.ylabel("Number of Visitors")
plt.title("Public Library Visitor Entry Time Distribution")
plt.xticks(range(0, 24))
plt.show()
```

```python
print("\nConclusion:")
print(f"The library experiences the highest visitor footfall around {peak_hour}:00 hours.")
print("This analysis helps in identifying peak usage hours for better staff and resource management.")
```
